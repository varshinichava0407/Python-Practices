```python
import random

def rand_number():
    return random.randint(1, 100)

def hard():
    print("Welcome to the Number Guessing Game!")
    print("I am thinking a number between 1 and 100")
    level_type = input("Choose a difficulty. Type 'easy' or 'hard': ")

    if level_type == 'easy':
        attempts = 8
    elif level_type == 'hard':
        attempts = 5
    else:
        print("Please type the correct level of difficulty")
       
    num = rand_number()
    print(f"You have {attempts} attempts to guess the number")

    for i in range(1, attempts + 1):
        guess = int(input("Make a guess: "))

        if guess == num:
            print(f"You got it!! The answer was {num}")
            return

        if guess < num - 20:
            print("Too less")
        elif num - 20 <= guess < num:
            print("Less but nearer")
        elif guess > num + 20:
            print("Too high")
        elif num < guess <= num + 20:
            print("High but nearer")

        remaining_attempts = attempts - i
        if remaining_attempts > 0:
            print(f"You have {remaining_attempts} attempts left. Guess again!\n")
        else:
            print(f"You ran out of attempts! The correct number was {num}")
```

```python
hard()
```
